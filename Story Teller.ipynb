```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load GPT2 model & tokenizer
print("Loading model...")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()
print("Model loaded!")

# Generate story from prompt
def generate_story(prompt, max_length=150):
    inputs = tokenizer.encode(prompt, return_tensors='pt')
    with torch.no_grad():
        outputs = model.generate(
            inputs,
            max_length=max_length,
            do_sample=True,
            top_k=50,
            top_p=0.95,
            temperature=0.9,
            num_return_sequences=1,
            pad_token_id=tokenizer.eos_token_id
        )
    story = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return story[len(prompt):]

# Chat loop
def chat():
    print("\nðŸ¤– Hello! I'm your storytelling companion.")
    print("ðŸ“– Tell me the kind of story you want (fantasy, sci-fi, horror...)")

    while True:
        user_input = input("\nYou: ")
        if user_input.lower() in ['exit', 'quit']:
            print("Goodbye! Come back for more stories!")
            break
        print("\nLet me think...")
        story = generate_story(user_input)
        print(f"\n{story.strip()}")

# Start chatbot
if __name__ == "__main__":
    chat()
```
